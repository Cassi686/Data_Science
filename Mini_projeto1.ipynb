```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as go 
import pandas as pd
import numpy as np
```

```python
# Importando Dataframe
mp = pd.read_csv('/content/kc_house_data[1].csv')
```

```python
#Visualizando dataframe
mp
```

```python
#Tratando coluna bathroons
mp['bathrooms'] = mp['bathrooms'].astype(int)
```

```python
#Tratando coluna date
mp['date'] = mp['date'].apply(lambda x : x[:-7])
```

```python
#Visualizando mudanças no dataframe
mp
```

```python
#Instalando biblioteca
!pip install pyzipcode
```

```python
#Importando biblioteca
from pyzipcode import ZipCodeDatabase
```

```python
# Criando variável
zcdb = ZipCodeDatabase()
```

```python
#Criando coluna city e visualizando mudança no dataframe
city = []
for zipcode in mp['zipcode']:
  try:
    zc=zcdb[zipcode]
    city.append(zc.city)
  except KeyError:
    if zipcode == 98077:
        city.append(str('Woodinville'))
    else:
        city.append(str('nan'))

mp['city'] = city
mp
```

```python
# Conferindo dados nulos na coluna city(nova coluna)
pd.isnull(mp['city'])
```

```python
# Criando nova coluna, valor por pés quadrados(medida americana)
sqft_total= mp['sqft_living'] + mp['sqft_lot']
sqft_valor=mp['price'] / sqft_total
mp['sqft_valor'] = sqft_valor
mp
```

```python
#Conferindo todas as colunas do dataframe
mp.columns
```

```python
#Retirando do dataframe colunas irrelevantes para a pesquisa
mp= mp.drop(['waterfront','view','yr_built','yr_renovated','lat','long','sqft_living15','sqft_lot15'],axis=1)
```

```python
# Criando novo dataframe(reincidence_mp),baseado no dataframe inicial. Criando coluna valorization no novo dataframe.
ids = mp['id']
reincidence_mp=mp[ids.isin(ids[ids.duplicated()])].sort_values('id')
reincidence_mp=reincidence_mp[['id','date','price','bedrooms','bathrooms','sqft_living','sqft_lot','sqft_above','sqft_basement','sqft_valor','floors','condition','grade','city']]
reincidence_mp=reincidence_mp.sort_values(['id','date']).reset_index(drop=True)
reincidence_mp['valorization']=np.zeros(reincidence_mp.shape[0])

```

```python
#Comparando colunas dos dois dataframes
print(mp.columns)
print(reincidence_mp.columns)
```

```python
#Tratando coluna data
reincidence_mp["date"]= pd.to_datetime(reincidence_mp["date"])
```

```python
#Criando coluna hold_time no novo dataframe
for index in range(1,int(reincidence_mp.shape[0])):
  if reincidence_mp.iloc[index]['id'] == reincidence_mp.iloc[index-1]['id']:
      reincidence_mp.at[index,'valorization'] = reincidence_mp.iloc[index]['price'] - reincidence_mp.iloc[index-1]['price']
      reincidence_mp.at[index,'hold_time'] = (reincidence_mp.iloc[index]['date'] - reincidence_mp.iloc[index-1]['date']).days
      
      
```

```python
#Criando colunas hold_time_year e valorization_year no novo dataframe

reincidence_mp['hold_time_year'] = reincidence_mp['hold_time']/365
reincidence_mp['valorization_year'] = reincidence_mp['valorization'] / reincidence_mp['hold_time_year']
reincidence_mp
```

```python
#Tratando novoas colunas
reincidence_mp['hold_time']=reincidence_mp['hold_time'].fillna(0)
reincidence_mp['hold_time_year']=reincidence_mp['hold_time_year'].fillna(0)
reincidence_mp['valorization_year']=reincidence_mp['valorization_year'].fillna(0)
```

```python
#visualizando novo dataframe
reincidence_mp
```

```python
#Conferindo dados nulos
reincidence_mp.isnull().sum()
```

```python
#Criando dataframe com os 10 imóveis que mais valorizaram
top10_valorization1=reincidence_mp.sort_values(by='valorization', ascending=False).head(10)
```

```python
#Visualizando dataframe com 10 imóveis que mais valorizaram
top10_valorization1
```

```python
#Criando dataframe com os 10 imóveis que menos valorizaram
tail10_valorization=reincidence_mp.sort_values(by='valorization', ascending=False).tail(10)
#Visualizando o dataframe com 10 imóveis que menos valorizaram
tail10_valorization
```

```python
#Criando gráfico: House price averange appreciation per city
# Gráfico de valorização média do preço das casas por cidade.
valorization_areas_year=reincidence_mp.groupby('city')['valorization_year']\
                        .mean().reset_index()\
                        .sort_values('valorization_year',inplace=False)\
                        .reset_index(drop=True)

valorization_areas=reincidence_mp.groupby('city')['valorization']\
                        .mean().reset_index()\
                        .sort_values('valorization',inplace=False)\
                        .reset_index(drop=True)

imobile_count=reincidence_mp.groupby('city')['id'].nunique().to_frame('imobile_count')
resale_info=pd.merge(valorization_areas_year,valorization_areas, on='city',how='inner')
resale_info=pd.merge(resale_info,imobile_count,on='city',how='inner')

resale_info_fig=px.bar(resale_info,x='city', y='valorization', color='imobile_count',
                      title='House price average appreciation per city')

resale_info_fig.add_trace(go.Scatter(x=resale_info['city'],
                                     y=resale_info['valorization_year'],
                                     name='ROI/Year',
                                     mode='lines'))

resale_info_fig.update_layout(title_x=0.5,
                              template='simple_white',
                              yaxis_title='',
                              xaxis_title='',
                              hovermode="x unified")

resale_info_fig.show()
```

```python
#Criando gráfico: Seattle ROI over selling
# Gráfico de retorno de investimento sobre venda em Seattle(cidade com maior movimentação de imóveis)
seattle_roi=reincidence_mp[(reincidence_mp['city']=='Seattle') & reincidence_mp['valorization']!=0].sort_values('date')

seattle_valorization_fig=px.line(seattle_roi,x='date',y='valorization',title='Seattle ROI over selling')
seattle_valorization_fig.add_trace(go.Scatter(x=seattle_roi['date'],
                                              y=seattle_roi['valorization_year'],
                                              name='ROI/Year',mode='lines'))
seattle_valorization_fig.update_layout(title_x=0.5,
                                       template='simple_white',
                                       yaxis_title='',
                                       xaxis_title='',
                                       hovermode='x unified')
seattle_valorization_fig.show()
```
